<h2> Problema</h2>
<p text-align='justify'> Las enfermedades cardiovasculares son la principal causa de muerte en el mundo, y se calcula que cobran 17,9 millones de vidas al año (OMS). La enfermedad de las arterias coronarias es el tipo más común de enfermedad cardíaca y se produce debido a las obstrucciones (placa) desarrolladas en el interior de las arterias coronarias (vasos sanguíneos que alimentan los músculos del corazón). Con este objetivo en mente, se quiere utilizar las técnicas de machine learning para construir un modelo que permita predecir qué pacientes pueden estar en riesgo de padecer este tipo de cardiopatía.</p>

Finalmente, hacemos la división entre nuestra variable objetivo <code>class</code> y el resto de variables.
```python
Y = arterias_coronarias['class']
X = arterias_coronarias.drop(['class'], axis=1)

X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.2, random_state=8)
```

Con esto, procedemos a hacer todo el entrenamiento sobre el conjunto de $X_{\mathrm{train}},Y_{\mathrm{train}}$ y guardamos el test en la "caja fuerte" hasta que sea el momento de la validación.

<h2>Procesamiento</h2>

---

Lo primero que pensamos fue que, viendo los valores de las variables categóricas resultado del perfilamiento, pareciera que ellas pudiesen llevar un orden definido, de acuerdo a la gravedad de la sintomatología. Si esto es así, es beneficioso pues, en vez de usar <i>One Hot Encoding</i> (aumentando la dimensionalidad del conjunto de datos - y el tiempo de cómputo -), podríamos definir una codificación ordinal más natural. Veamos cada una de las variables con categorías: <code>cp</code>, <code>restecg</code>, <code>slope</code> y <code>thal</code>.

<ul>
    <li> <code>cp</code>: Esta variable está relacionada con el tipo de dolor de pecho. Puede tomar los valores: Typical Angina, Atypical Angina, Nonanginal pain y Asymptomatic. La primera es la de mayor gravedad, pues corresponde a la totalidad de los 3 síntomas. La segunda, corresponde a la presencia de 2 síntomas. La primera a uno sólo y, la última, a ninguno <a href="#angina">[3]</a>. Por lo tanto, vemos un orden natural. Decidimos codificar nuestra variable de la siguiente forma: 0 corresponde al asíntomático y, 3, a la Angina Típica.</li>
    <li> <code>restecg</code>: En cuanto a esta variable, que responde a un electrocardiograma en reposo, la peor condición corresponde a Left ventricular hypertrophy (esta será 2), seguida de ST-T wave abnormality (1) y, por último, la condición normal (0).</li>
    <li> <code>slope</code>: La condición normal corresponde a <i>upsloping</i>, pues es lo que debe verse para una persona haciendo ejercicio. A esta, le asignamos 0. La condición normal (1) y downsloping (2) son las que siguen, en dicho orden, en gravedad <a href="#slope">[4]</a>. </li>
    <li> <code>thal</code>: Esta variable está relacionada con un defecto. El mejor caso es "normal" (i.e. no tenerlo, a lo cual le asignamos un 0), seguido de reversible (1) y fixed (2) <a href="#thal">[5]</a>. </li>
</ul>

Con esto, procedemos a construir nuestro codificador ordinal.

```python
cp_ordenados = [['Asymptomatic', 'Nonanginal pain', 'Atypical Angina', 'Typical Angina']]
restecg_ordenados = [['Normal', 'Having ST-T wave abnormal', 'Left ventricular hypertrophy']]
slope_ordenados = [['Unsloping', 'Flat', 'Downsloping']]
thal_ordenados = [['Normal', 'Reversible defect', 'Fixed']]
```

```python
codificador_ordinal = make_column_transformer(
    (OrdinalEncoder(categories=cp_ordenados), ['cp']),
    (OrdinalEncoder(categories=restecg_ordenados), ['restecg']),
    (OrdinalEncoder(categories=slope_ordenados), ['slope']),
    (OrdinalEncoder(categories=thal_ordenados), ['thal']),
)
```

Es importante notar que el uso del imputer es innecesario pues no hay datos faltantes. Procedemos a seleccionar las columnas numéricas de las categóricas. A las numéricas, les haremos un escalamiento, en este caso dado por una normalización a 1. Esta es conveniente pues conserva la misma distribución de los datos pero los lleva a la misma escala, tal que no haya algunas variables que, espúremente, pesen más que otras al construir el modelo. Este escalamiento también lo hacemos a las categóricas después de haberlas codificado como ordinales.

```python
# Seleccionamos las columnas numericas.
columnas_numericas = list(X_train.select_dtypes(include=['int64', 'float64']).columns)
# Seleccionamos las columnas categoricas.
columnas_categoricas = list(X_train.select_dtypes(exclude=['int64', 'float64']).columns)

# Aplicamos el escalador normalizador.
transformador_numerico = Pipeline(steps=[
    ('scaler', MinMaxScaler())])

# Aplicamos el codificador ordinal y, posteriormente, el escalador normalizador.
transformador_categorico = Pipeline(steps=[
    ('encoder', codificador_ordinal),
    ('scaler', MinMaxScaler())])

#Creamos una pipeline con ambos preprocesadores
preprocesador_ordinal = ColumnTransformer(
    transformers=[
        ('num', transformador_numerico, columnas_numericas),
        ('cat', transformador_categorico, columnas_categoricas)])
```

En principio, pensamos que un buen modelo que se ajusta a nuestros datos es la regresión logística. Lo anterior, pues tenemos un problema de clasificación y queremos determinar la probabilidad de que un paciente tenga (1) o no (0) la enfermedad de arterias coronarias. El modelo propuesto funciona muy bien para estos casos. Más adelante, analizaremos otros modelos de clasificación.

Ya podemos crear la pipeline de preprocesamiento, como vemos a continuación.

```python
# Creamos la pipeline de pre-procesamiento para nuestro estudio preliminar.
pipeline_ordinal = Pipeline(steps=[('preprocesador', preprocesador_ordinal),
                                          ('classifier', LogisticRegression(solver='liblinear', penalty='l2'))])
```

Note que hemos usado la penalidad l2 (regresión Ridge), pues es la que usa la librería por defecto. Cuando hagamos validación cruzada, veremos si este es el método más eficiente o si, por el contrario, conviene usar mejor una regresión Lasso para la penalidad. Asimismo, hemos usado el solver liblinear. Lo anterior pues, de acuerdo a la documentación de <code>scikit-learn</code> esta es la que mejor se ajusta a conjuntos de datos pequeños como el nuestro.

Con esto, ya podemos hacer fit al modelo.

```python
# Hacemos fit al modelo
pipeline_ordinal.fit(X_train, Y_train)
# Obtenemos las predicciones para los datos de entrenamiento
pred_ord = pipeline_ordinal.predict(X_train)
# Observamos la matriz de confusión correspondiente a nuestra predicción
plot_confusion_matrix(pipeline_ordinal,X_train,Y_train,values_format='d')
plt.title('Matriz de Confusión usando el Clasificador Ordinal')
plt.show()
```

```python
# Asimismo, vemos el reporte del algoritmo
print('----------Reporte para el Modelo 1----------')
print(classification_report(Y_train,pred_ord))
```

Note que, en entrenamiento, parece haber tenido un muy buen desempeño nuestro clasificador. Evidencia de ello es que tanto la precisión, recall y f1-score promedio tuvieron excelentes resultados (0.85). No obstante, no basta con quedarnos con estos: debemos probar diversos tipos de codificación (como OneHot Encoding), escalamiento (No solo MinMax, sino estándar o 'passthrough' (i.e. que no haga ningún escalamiento, por ejemplo)) e hiperparámetros determinados por cada algoritmo estudiado.

Procedemos a construir el Pipeline para One-Hot Encoding, tal que tenga en cuenta todas las variables de nuestro problema.

```python
# Seleccionamos las columnas numericas.
columnas_numericas = list(X_train.select_dtypes(include=['int64', 'float64']).columns)
# Seleccionamos las columnas categoricas.
columnas_categoricas = list(X_train.select_dtypes(exclude=['int64', 'float64']).columns)

# Aplicamos el escalador normalizador.
transformador_numerico = Pipeline(steps=[
    ('scaler', MinMaxScaler())])

# Aplicamos el codificador ordinal y, posteriormente, el escalador normalizador.
transformador_categorico = Pipeline(steps=[
    ('encoder', OneHotEncoder())])


preprocessor_one_hot = ColumnTransformer(
    transformers=[
        ('num', transformador_numerico, columnas_numericas),
        ('cat', transformador_categorico, columnas_categoricas)])
```

<h2>Modelo por regresión logística</h2>

---

Con un modelo básico funcionando, procedemos a hacer un GridSearch junto con validación cruzada (n=5), de manera que una búsqueda más exhaustiva sobre los hiperparámetros nos provea un mejor modelo, y con la validación cruzada, resultados estadísticamente más significativos. Así las cosas, para la normalización elegimos cuatro opciones: no normalizar <code>(passthrough)</code>, escalar con el máximo valor absoluto <code>MaxAbsScaler</code>, con la desviación estándar <code>StandardScaler</code> y con el rango (<code>MinMaxScaler</code>). Para C (el parámetro inverso de la regularización), se hizo un barrido de veinte valores en el espacio logarítmico entre $10^{-3}$ y $10^{3}$. Entre mayor sea el valor de C, más alta es la regularización del modelo. Asimismo, usamos dos tipos de penalización: L1 y L2. De estos dos tipos de penalización, que corresponden a Lasso y Ridge respectivamente, queremos ver cuál es más óptimo para nuestro modelo. Note que el solver <code>liblinear</code> soporta ambos tipos de penalización.


```python
# Definimos los estimadores de nuestra Pipeline original. 
# Note que iniciamos con el codificador ordinal definido.
estimators = [('encoder', codificador_ordinal),
              ('normalize',MinMaxScaler()),
              ('clf', LogisticRegression(solver='liblinear')),
              ]
pipe = Pipeline(estimators)

# Evaluamos tanto nuestro codificador ordinal como el que hicimos con OneHot.
param_grid = dict(encoder=[preprocesador_ordinal, preprocessor_one_hot],
                  normalize=['passthrough', MaxAbsScaler(), StandardScaler(), MinMaxScaler()],
                  clf__C = np.logspace(-3,3,20),
                  clf__penalty = ['l1', 'l2'],
                 )
                 

grid_search = GridSearchCV(pipe, param_grid=param_grid,cv=5,verbose=1,scoring='accuracy')

grid_search.fit(X_train,Y_train.ravel())
```

El mejor modelo encontrado uso los parámetros mostrados a continuación:

```python
print(grid_search.best_params_)
```

Para nuestra sorpresa, la mejor combinación de hiperparámetros incluye el uso del one-hot encoding para las variables categóricas. Por otro lado, el modelo prefirió el escalamiento de las variables numéricas usando MinMaxScaling() mientras que, para las categóricas, prefirió el uso del escalador estándar. 

Asimismo, note que se escogió la penalidad l2 (correspondiente a la regresión Ridge, lo cual tiene sentido pues es la que está implementada por defecto en la librería). Es posible ver que el valor de C es pequeño: 0.009 aproximadamente, lo que nos dice que el modelo tiene poca regularización. Podemos ver que valió la pena ajustar este parámetro pues, por defecto de la librería, está en 1; y, en realidad, nuestro modelo usó uno mucho más pequeño.


Veamos ahora cómo le va a nuestro modelo en su predicción con el test set, que ya podemos usar para validar.

```python
dump(grid_search.best_estimator_, 'modelo_regresion_logistica.joblib') 

logistic_model = load('modelo_regresion_logistica.joblib') 

pred1 = logistic_model.predict(X_train)
print('----------Reporte para el Mejor Modelo LR----------')
print(classification_report(Y_train,pred1))
```

Aquí es posible evidenciar que hubo una mejora (pequeña) en el modelo en cuanto a entrenamiento, con respecto a nuestro intento inicial descrito más arriba. Vemos una muy ligera mejora sobre el modelo probado inicialmente sin la ayuda del GridSearch. Note que la codificación con OneHotEncoding da como resultado un f1-score con 0.02 puntos más que aquella con el clasificador ordinal que diseñamos (a pesar de que tuvo un buen desempeño inicial). Lo anterior pues, además, nos estamos cerciorando de utilizar los mejores hiperparámetros de la regresión logística, usando validación cruzada.

<h2>Modelo por KNN</h2>

---

Ahora procedemos a probar otros modelos, en particular KNN. 
KNN es un algoritmo de clasificación que permite determinar la clase de un dato de acuerdo a la información de sus K-vecinos. Para ello procedemos de forma similar a construir un Pipeline apropiado. En este caso usamos los mismos escaladores que para el caso de Regresión Logística y los mismos tipos de codificación: el ordinal (diseñado por nosotros) y OneHotEncoding.

Asimismo, hacemos una búsqueda de los hiperparámetros propios de KNN. Esta se hace sobre tres elementos: el número de vecinos usados en el modelo (barrimos sobre los números impares entre tres y once), el peso de cada vecino (uniformes o pesando con el inverso de la distancia, tal que "el voto" de los vecinos más cercanos pese más que el de los más lejanos), y el parámetro p, que es el exponente de la norma para la distancia y que define la métrica. Para p usamos 1, 2 y 3, que corresponden respectivamente a métrica de Manhattan (o del taxista), euclidiana (la estándar que entendemos por distancia en espacios no-curvos), y una general de Minkowski con exponente igual a tres. 

```python
# KNN
estimators = [('encoder', codificador_ordinal),
              ('normalize',MinMaxScaler()),
              ('clf', KNeighborsClassifier()),
              ]
pipe = Pipeline(estimators)

param_grid = dict(encoder=[codificador_ordinal, preprocessor_one_hot],
                  normalize=['passthrough', MaxAbsScaler(), StandardScaler(), MinMaxScaler()],
                  clf__n_neighbors = [3, 5, 7, 9, 11],
                  clf__weights = ['uniform', 'distance'],
                  clf__p = [1,2,3]
                 )

grid_search_knn = GridSearchCV(pipe, param_grid=param_grid,cv=5,verbose=1,scoring='accuracy')
grid_search_knn.fit(X_train,Y_train.ravel())
```

Procedemos a ver cuáles fueron los hiperparámetros óptimos.

```python
print(grid_search_knn.best_params_)
```

La mejor combinación de hiperparámetros encontrada para KNN implica usar el MinMaxScaler para las variables numéricas y el One-Hot encoder para las categóricas. Sobre el modelo en sí mismo, el mejor modelo usa pesos uniformes sobre nueve vecinos, usando una métrica de Minkowski con p=3. Los puntajes sobre la validación cruzada (n=5) se muestran a continuación.

```python
dump(grid_search_knn.best_estimator_, 'modelo_knn.joblib') 

knn_model = load('modelo_knn.joblib') 

pred_knn = knn_model.predict(X_train)
print('----------Reporte para el Mejor Modelo KNN----------')
print(classification_report(Y_train,pred_knn))
```

Los puntajes obtenidos son perfectos. Tanto la precisión y la exhaustividad, como la métrica f1 resultado de los dos, nos da una exactitud del 100% en la clasificación. En la mayoría de modelos, esto es un problema puesto que significa que el modelo está haciendo overfitting. Esto podría estar ocurriendo debido a la poca cantidad de datos de nuestro training set. No obstante, no sabemos esto hasta la validación.  


<h2>Modelo por SVM</h2>

---

Finalmente podemos expĺorar el uso de máquinas de vectores de soporte para la clasificación. Probamos los mismos cuatro escaladores que usamos para la regresión logística y para el SVM en este caso, para normalizar el conjunto de datos, así como ambas codificaciones: la ordinal (nuestra) y la OneHot. Para la máquina de vectores de soporte tenemos más hiperparámetros propios del algoritmo sobre los cuales hacer el GridSearch: lo primero es el kernel a utilizar, donde probamos el lineal, el polinómico y el de función de base radial (rbf). En el caso del kernel polinómico probamos sobre grados 2, 3, 4 y 5. Para el de RBF y el polinomial usamos como coeficientes el tipo auto y el escalado (el segundo es igual al primero, $1/n$, dividido sobre la varianza del conjunto de datos).  Finalmente, para el parámetro de regularización C hicimos un barrido de siete valores en el espacio logarítmico entre $10^{-3}$ y $10^{3}$. Al igual que para la LogReg y el modelo SVM hacemos una validación cruzada con n=5.

<b>NOTA IMPORTANTE: </b> El espacio de búsqueda de esta validación cruzada es mucho mayor que el de los anteriores por lo que, al correr la celda siguiente, note que puede demorarse aproximadamente <b>6-8</b> minutos.

```python
# SVM
estimators = [('encoder', codificador_ordinal),
              ('normalize',MinMaxScaler()),
              ('clf', SVC()),
              ]
pipe = Pipeline(estimators)

param_grid = dict(encoder=[preprocesador_ordinal, preprocessor_one_hot],
                  normalize=['passthrough', MaxAbsScaler(), StandardScaler(), MinMaxScaler()],
                  clf__C = np.logspace(-3,3,7),
                  clf__kernel = ['linear', 'poly', 'rbf'],
                  clf__degree = [2, 3, 4, 5],
                  clf__gamma = ['auto', 'scale']
                 )

grid_search_svm = GridSearchCV(pipe, param_grid=param_grid,cv=5,verbose=1,scoring='accuracy')
grid_search_svm.fit(X_train,Y_train.ravel())
```

El mejor conjunto de parámetros y los puntajes sobre la validación cruzada se muestran a continuación

```python
print(grid_search_svm.best_params_)
```

Note que el mejor que, de nuevo, los codificadores y escaladores elegidos son los mismos de todos los modelos anteriores. Este es un argumento suficientemente fuerte para sostener que, dados los datos y features del modelo, lo mejor es usar OneHotEncoder() para las variables categóricas y, para las numéricas, un escalador que normalice a la unidad, MinMaxScaler(). 

Note que, en cuanto a los hiperparámetros de SVM, el valor de C es bastante bajo (es decir, hay poca regularización), comparado con el valor por defecto de la librería (1.0). Asimismo, note que se escogió el kernel polinomial (en particular, se usó una función de grado 2).

Veamos los parámetros de entrenamiento,

```python
dump(grid_search_svm.best_estimator_, 'my_model_svm.joblib') 

svm_model = load('my_model_svm.joblib') 

pred_svm = svm_model.predict(X_train)
print('----------Reporte para el Mejor Modelo----------')
print(classification_report(Y_train,pred_svm))
```

Para nuestra sorpresa, el puntaje alcanzado para el mejor modelo de SVM es inferior a los de regresión logística y de KNN. Una posibilidad es que el modelo es muy complejo para la limitada cantidad de datos con que contamos, o simplemente que esa familia de modelos no se ajusta tan bien al problema en cuestión y sus peculiaridades. 


De acuerdo a todo lo visto previamente y después de haber entrenado tres algoritmos de clasificación diferentes, debemos elegir cual de los tres se adaptaría mejor a la validación. Descartamos, primero, el modelo entrenado inicialmente (sin validación cruzada) así como SVM, pues minimizan la precisión, exhaustividad y, por tanto, el puntaje f1 del modelo. Ahora, podemos elegir entre KNN y Regresión Logística. El modelo de regresión logística indica claramente que no se está haciendo ni overfitting ni underfitting, pues no es de 1.0 (como sería el caso superior, tal que sospecháramos esta condición), pero tampoco es demasiado bajo, tal que el modelo está aprendiendo a un buen ritmo. El modelo de KNN tiene resultados sospechosamente excelentes. Debido a que en este caso podemos probar el desempeño de ambos en validación, miraremos cual de los dos se desempeña mejor para tomar decisiones concluyentes. No obstante, en un caso donde simplemente tuviésemos que desplegar el modelo sin tener acceso a los datos de test antes, sería más conveniente elegir el modelo de regresión logística: pues de este, no sospechamos ni overfitting ni underfitting.
